```python
import os

from ultralytics import YOLO
import cv2
```

```python
IMG_DIR = os.path.join('.', 'images')

img_path = os.path.join(IMG_DIR, 'test.jpg')
img_path_out = '{}_out.jpg'.format(img_path)

img = cv2.imread(img_path)
H, W, _ = img.shape
out = cv2.imwrite(img_path_out, img)
```

```python
model_path = os.path.join('.', 'runs', 'detect', 'train0002', 'weights', 'last.pt')

model = YOLO(model_path)

threshold = 0.5

class_name_dict = {0: 'plate_blue',1: 'plate_yellow'}
```

```python
results = model(img)[0]

for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result

    if score > threshold:
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
        cv2.putText(img, class_name_dict[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

cv2.imwrite(img_path_out, img)
```
